```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import random
import collections
import numpy as np
import pandas as pd
import datetime
import random
import scipy.stats as cov
```

```python
cas = pd.read_csv("dataset.csv")
```

```python
def plot_discrete_continuous(col):
    df_discrete = pd.pivot_table(cas, index=col,values='id', columns='Success', aggfunc=np.count_nonzero)
    #df_discrete[0] = np.sqrt(df_discrete[0])
    df_discrete.plot(kind="bar", figsize=(20,15))
```

```python
plot_discrete_continuous("Age")
```

```python
plot_discrete_continuous("property_type")
```
